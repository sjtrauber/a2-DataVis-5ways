```python
import pandas as pd

url = "https://raw.githubusercontent.com/cs573-22s/a2-DataVis-5ways/main/cars-sample.csv"
data = pd.read_csv(url)

data.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})

x = data.Weight
y = data.MPG
z = data.Manufacturer

colors = {'ford':'red', 'toyota':'green', 'bmw':'blue', 'honda':'yellow', 'mercedes':'pink'}

plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.scatter(x,y, alpha = 0.5, sizes = x/10, color = data['Manufacturer'].map(colors))
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.xticks(np.arange(2000, max(x)+500, 1000))
plt.yticks(np.arange(10, max(y), 10))
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(data, x="Weight", y="MPG", color = "Manufacturer", size = "Weight", opacity = 0.5)
fig.update_layout(xaxis = dict(tickmode = 'linear', dtick = 1000), yaxis = dict(tickmode = 'linear', dtick = 10))
fig.show()
```
